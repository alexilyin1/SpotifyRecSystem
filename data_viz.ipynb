```python
for ind, col in enumerate(dat.columns):
    if dtypes_dict[col] == 'float64':
        plt.figure()
        plt.hist(
            dat[col], 
            color=random.choice(list(colors.TABLEAU_COLORS.values()))
        )
        plt.title(col)
```
